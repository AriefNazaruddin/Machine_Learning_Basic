```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
df.columns
```

```python
plt.scatter(df.Mileage, df['Sell Price($)'], color='red', marker='.')
```

```python
temp_df = pd.get_dummies(df['Car Model'])
df_dump = temp_df.astype(int)
df_dump
```

```python
merged = pd.concat([df, df_dump], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'], axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
model.fit(X,y)
```

```python
model.score(X, y)
```
