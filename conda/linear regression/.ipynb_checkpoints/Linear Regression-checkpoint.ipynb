```python
# Linear resgression is a statistical method used to model the relationship between dependent variable and one or more independent variables by fitting
# a linear equation to observed data. It is one of the simplest and most commonly used techniques in statistical modelling and machine learning for regression taks
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(RM)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[5500]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767123*5500+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d
```

```python
d.to_csv("prediction.csv", index=False)
```
